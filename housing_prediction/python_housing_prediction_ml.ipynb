## import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
import numpy as np

```python
# Read the housing.csv file into pandas dataframe
housing_df = pd.read_csv('housing.csv')

# Check feature vector column name, count and datatype
housing_df.info()
```

```python
plt.hist(housing_df['median_house_value'], bins = 80)
plt.xlabel('House Values') 

# Most houses price range from 100k to 200k with few outliers
```

```python
housing_df.hist(bins = 80, figsize=(20,15))
```

```python
# creating correlation matrix 
corr = housing_df.corr(numeric_only = True)
print(corr)
```

```python
# creating visual heatmap

plt.figure(figsize = (8,8))

sns.heatmap(corr, annot=True)
plt.show()
```
